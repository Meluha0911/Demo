# Method of Pandas

```python
# 1.append
# 2.concat
# 3.merge
```

```python
# Append Function
# -- In pandas append function is like adding one table(DF) to the bottom of another table(DF)
# -- It combines data from two tables vertically, stacking one on the top of other
```

```python
# create 1st DF
data1 = {'Name':['Alice','Bob'],'Age':[25,30]}
df1 = pd.DataFrame(data1)
df1
```

```python
# create 2nd DF
data2 = {'Name':['Charlie','David'],'Age':[22,35]}
df2 = pd.DataFrame(data2)
df2
```

```python
# Append second DF to the First DF
combined_df = pd.concat([df1,df2],ignore_index=True)
```

```python
combined_df
```

```python
# Merge Method
# Merge is use to combine the Two Df -- base on the common column'
data1 = {'StudentID':[1,2,3],'Name':['Alice','Bob','Charlie']}
student_df = pd.DataFrame(data1)
student_df
```

```python
data2 = {'StudentID':[2,3,4],'Score':[85,92,78]}
scores_df = pd.DataFrame(data2)
scores_df
```

```python
# Merge the Two DF based on the 'StudentID' column''
merged_df = pd.merge(student_df,scores_df,on='StudentID')
merged_df
```

```python
# Join in The Pandas
df1 = pd.DataFrame({'A':[1,2,3]},index=['a','b','c'])
df2 = pd.DataFrame({'B':[4,5,6]},index=['a','b','d'])
result = df1.join(df2,how='left')
```

```python
result
```

```python
result = df1.join(df2,how='right')
result
```

```python
result = df1.join(df2,how='outer')
result
```

```python
# Joining the numpy array's --> means putting the content of two or more array's in the single 
# array
# In SQL we join tables by using the key while in numpy we join the array by using the axis both 
# are same
# we can use concat function for that purpose -- along with the axis -- by default the axis is 0
```

```python
# Joining the two_array
na1 = np.array([1,2,3])
na2 = np.array([4,5,6])
na3 = np.concatenate((na1,na2))
na3
```

```python
# Concatating / joining the 2 2D array's along with columns (axis = 1)
na1 = np.array([[1,2],[3,4]])
na2 = np.array([[5,6],[7,8]])
na3 = np.concatenate((na1,na2),axis=1)
na3
```

```python
# Concatnating / joining 2 2d array's along rows -- use axis = 0
na1 = np.array([[1,2],[3,4]])
na2 = np.array([[5,6],[7,8]])
na3 = np.concatenate((na1,na2),axis=0)
na3
```

```python
# Joining the arrys using the stack function -- stack function is same as concatenate function
# the only diff is that stacking is done along the new axis
na1 = np.array([1,2,3])
na2 = np.array([4,5,6])
na3 = np.stack((na1,na2),axis = 1)
na3
```

```python
# Different types of stack -- there are two types of stack -- stack along row's and the stacl alng
# column's
na1 = np.array([1,2,3])
na2 = np.array([4,5,6])
na3 = np.hstack((na1,na2)) # hstack is used for stacking via row
na4 = np.vstack((na1,na2)) # vstack is used for stacking via column's
na5 = np.dstack((na1,na2)) # dstack is used for stackign along the height
print(na3)
print(na4)
print(na5)
```

```python
# Splitting the Numpy array -- How to do splitting -- splitting is the reverse of the joining
# splitting break the one array into multiple array's
# we will use array_split() function for that purpose
na = np.array([1,2,3,4,5,6])
na_new = np.array_split(na,2)
na_new
```

```python
# How to add label -- means applying labels to the x and y axis -- for that use label functionality
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y)
plt.xlabel('language',fontsize=15,color='r') # giving the label to x_axis
plt.ylabel('popularity',fontsize=15,color='r') # giving the label to y_axis
plt.title("Programming Language") # Giving the title to graph
plt.plot()
```

```python
# How to reduce/increase width of the bar -- use width functionality for that purpose
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,width=0.8) # scale of width ranges from 0.0 to 1
plt.show()
```

```python
# How to get python to the edge and not to the center -- bringing the name of x_axis to the edges
# -- use align parameter for that purpose
x = ['python','c','c++','java']
y = [85,70,60,82]
plt.bar(x,y,align='edge')
plt.show()
# Inside align you can pass two parameter i.e edge and center(by default)
```
